```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 삼성전자 일별 시세 데이터

def Samsung_daily_price(page_no):
    url = f"https://finance.naver.com/item/sise_day.nhn?code=005930&page={page_no}"
    headers = {"user-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36"}
    response = requests.get(url,headers=headers)
    html = bs(response.text,"lxml")
    temp = html.select("table")
    data = pd.read_html(str(temp))
    data = data[0].dropna()
    
    return data

Samsung_daily_price(1).to_csv("samsung.csv",encoding="utf-8")

```
